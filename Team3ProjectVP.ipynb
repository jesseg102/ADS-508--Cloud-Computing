```python
import pandas as pd
import boto3
import matplotlib.pyplot as plt
import seaborn as sns

# Initialize S3 client
s3_client = boto3.client('s3')

# Load vehicle_speed.csv
csv_bucket = 'vpproject'
csv_key = 'vehicle_speed/vehicle_speed.csv'
csv_object = s3_client.get_object(Bucket=csv_bucket, Key=csv_key)
vehicle_speed_df = pd.read_csv(csv_object['Body'])

# Display basic information about the DataFrame
print("Vehicle Speed DataFrame:")
print(vehicle_speed_df.info())
print()

# Display summary statistics
print("Summary Statistics:")
print(vehicle_speed_df.describe())
print()

# Visualize distributions of numerical variables
sns.pairplot(vehicle_speed_df)
plt.show()

# Visualize correlation matrix
correlation_matrix = vehicle_speed_df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

```python
import pandas as pd
import boto3
import matplotlib.pyplot as plt
import seaborn as sns

# Initialize S3 client
s3_client = boto3.client('s3')

# Load vehicle_speed.csv
csv_bucket = 'vpproject'
csv_key = 'vehicle_speed/vehicle_speed.csv'
csv_object = s3_client.get_object(Bucket=csv_bucket, Key=csv_key)
vehicle_speed_df = pd.read_csv(csv_object['Body'])

# Display basic information about the DataFrame
print("Vehicle Speed DataFrame:")
print(vehicle_speed_df.info())
print()

# Display summary statistics
print("Summary Statistics:")
print(vehicle_speed_df.describe())
print()

# Visualize histograms of numerical variables
vehicle_speed_df.hist(figsize=(10, 8))
plt.tight_layout()
plt.show()

# Visualize scatter plots of numerical variables
sns.pairplot(vehicle_speed_df)
plt.show()

# Visualize box plots of numerical variables grouped by categorical variables
plt.figure(figsize=(12, 8))
sns.boxplot(x='way_type', y='target_speed', data=vehicle_speed_df)
plt.xlabel('Way Type')
plt.ylabel('Target Speed')
plt.title('Distribution of Target Speed by Way Type')
plt.show()

# Visualize heatmap of correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(vehicle_speed_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```
